```python
def set_cosmo_params(bestfit):
    
    Cosmo_params = model.CAMBparams()
    
    if 'w0_fld,' in bestfit.keys():
        w0=float(bestfit['w0_fld,'])
    else:
        w0=-1
    
    Cosmo_params = camb.set_params(
        H0=float(bestfit['H0,']),
        ombh2=float(bestfit['omega_b,']),
        omch2=float(bestfit['omega_cdm,']),
        tau=float(bestfit['tau_reio,']),
        ns=float(bestfit['n_s,']),
        As=np.exp(bestfit['ln10^{10}A_s,'])*(10**-10),
        w=w0,
        omk=0,
        mnu=0.06,
        nnu=3.044,
    )
    
    return Cosmo_params
```

```python
def get_distance_moduli(bestfit, z, verbose=True):
    
    params=set_cosmo_params(bestfit)
    
    mu_b=float(bestfit['M,'])
    
    if verbose:
        print('Setting the following best-fit parameters:')
        print('w=',params.DarkEnergy.w)
        print('H0=',params.H0)
        print('ombh2=',params.ombh2)
        print('omch2=',params.omch2)
        print('tau=',params.Reion.optical_depth)
        print('ns=',params.InitPower.ns)
        print('As=',params.InitPower.As)
        print('M=',mu_b)
    
    results=camb.get_results(params)

    DA = results.angular_diameter_distance(z)
    DL = DA*(1 + z)**2
    distance_moduli=5*np.log10(DL)+25+mu_b

    if verbose:
        print('Returning the distance moduli \n')
    
    return distance_moduli
```
