```python
from __future__ import print_function
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import sys, os
sys.path.insert(0,os.path.realpath(os.path.join(os.getcwd(),'..')))
from getdist import plots, loadMCSamples, MCSamples
import getdist
# use this *after* importing getdist if you want to use interactive plots
# %matplotlib notebook
import matplotlib.pyplot as plt
import IPython
#print('GetDist Version: %s, Matplotlib version: %s'%(getdist.__version__, plt.matplotlib.__version__))
#matplotlib 2 may not work very well without usetex on, can uncomment
#plt.rcParams['text.usetex']=True
from matplotlib import rc
import seaborn as sns
from scipy.constants import c
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import argparse
import matplotlib.gridspec as gridspec
```

```python
#delensed roots to chains

full_camspec    =  './chains/ms_des/cam_ms_des/chains/musigma_des'
full_hillipop   =  './chains/ms_des/hil_ms_des/chains/musigma_des'

camspec=getdist.loadMCSamples(full_camspec, settings={'ignore_rows':0.3,'smooth_scale_1D': 0.9,'smooth_scale_2D': 0.9});
hillipop=getdist.loadMCSamples(full_hillipop, settings={'ignore_rows':0.3,'smooth_scale_1D': 0.9,'smooth_scale_2D': 0.9});
```

```python
#delensed roots to chains

TT_camspec    =  './chains/ms_des/camt_ms_des/chains/musigma_des'
TT_hillipop   =  './chains/ms_des/hilt_ms_des/chains/musigma_des'

camspec_TT=getdist.loadMCSamples(TT_camspec, settings={'ignore_rows':0.3,'smooth_scale_1D': 0.9,'smooth_scale_2D': 0.9});
hillipop_TT=getdist.loadMCSamples(TT_hillipop, settings={'ignore_rows':0.3,'smooth_scale_1D': 0.9,'smooth_scale_2D': 0.9});
```

```python
g = plots.get_subplot_plotter(width_inch=5)
g.settings.alpha_filled_add = 0.6
from matplotlib import cm

upper_kwargs = {'contour_colors': ['cadetblue'], 'contour_ls': ['-'], 
                'filled': [True], 'show_1d': [True], 'contour_lws':[1]}


g.triangle_plot([camspec], ['S8','sigma0','mu0'], 
                contour_colors=['navy'], filled=[True], markers={'mu0':0 ,'sigma0':0 },
                upper_roots = [camspec_TT],
                contour_ls=['-','--'],
                upper_kwargs = upper_kwargs, 
                upper_label_right=True,
                legend_labels=['Camspec','Camspec (TT-only)']);
g.export('sigma_Camspec.pdf')
```

```python
g = plots.get_subplot_plotter(width_inch=5)
from matplotlib import cm

upper_kwargs = {'contour_colors': ['olive'], 'contour_ls': ['-'], 
                'filled': [True], 'show_1d': [True], 'contour_lws':[1]}


g.triangle_plot([hillipop], ['S8','sigma0','mu0'], 
                contour_colors=['darkgreen'], filled=[True], markers={'mu0':0 ,'sigma0':0 },
                upper_roots = [hillipop_TT],
                contour_ls=['-','--'],
                upper_kwargs = upper_kwargs, 
                upper_label_right=True,
                legend_labels=['HiLLiPoP','HiLLiPoP (TT-only)']);
g.export('sigma_Hillipop.pdf')
```
