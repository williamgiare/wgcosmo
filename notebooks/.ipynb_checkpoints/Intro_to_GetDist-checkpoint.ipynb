### Step 3: Reading MCMC chains

- We have some chains in the repository `./data/Planck/chains` produced by using Planck-2018 CDM data with Cobaya 

- In the folder `chains`, there are many subfolders, each for a different model of cosmology (i.e., minimal extensions compared to $\Lambda$CDM).

- In each folder, there are a number of files produced in output by `Cobaya`:
  - `file.n.txt` are the n MCMC chains that we want to analyze (here n=1...4).
  - `file.covmat` is the covariance matrix among parameters.
  - `file.input.yaml` is a YAML file summarizing the original YAML file used to run `Cobaya`.
  - `file.updated.yaml` is a YAML file updating the original YAML file used to run `Cobaya` (i.e., the same file with more info).
  - `file.progress` contains information about the progress in chain convergence.


We now read the chains for the $\Lambda$CDM model using `Getdist`

- the chains are in the folder `./data/Planck/chains/lcdm`
- the root of the chains is 'lcdm', we we have the following files:
  - `lcdm.1.txt`
  - `lcdm.2.txt`
  - `lcdm.3.txt`
  - `lcdm.4.txt`
  - `lcdm.covmat` 
  - `lcdm.input.yaml` 
  - `lcdm.updated.yaml` 
  - `lcdm.progress` 
  
Notice that we will read all the 4 chains from `lcdm.1.txt` to `lcdm.4.txt` as follows:

```python
path_to_chain='../data/Planck_chains/lcdm/lcdm'
```

NOTE: the last term provided in `path_to_chain`  should be the root of the chains without `.n.txt`

```python
samples = getdist.loadMCSamples(path_to_chain, 
                                settings={'ignore_rows':0.3} #other settings can be given here
                               )
```

NOTE: the command `'ignore_rows': 0.3` is used to remove the first 30% of points in the chains corresponding to the burn-in phase (see notes on statistical methods).

### Step 4: Testing the convergence of the Chains

Samples now contain all the information on the chains. We can use the functions in GetDist to access several pieces of information.

For instance, to test that the Gelman-Rubin criterion for chain convergence is satisfied (i.e., $R-1\ll 1$), using the function in the module `getGelmanRubin()`, simply use: `samples.getGelmanRubin()`.


```python
print('R-1=',samples.getGelmanRubin())
```

### Step 5: Constraints on cosmological parameters

We can obtain the numerical constraints on cosmological parameters at different statistical levels using `samples.getInlineLatex('parameter_name', 'limit')`, where:

- `parameter_name` should be the name of the cosmological parameter, consistent with your definition in the original YAML file. 

   - For the standard six parameters, the CAMB nomenclature would be {`As`, `ns`, `ombh2`, `omch2`, `tau`, `H0` (or equivalently `theta_MC_100`)}.
   - For the standard six parameters, the CLASS nomenclature would be {`A_s`, `n_s`, `omega_b`, `omega_cdm`, `tau_reio`, `H0` (or equivalently `theta_s_100`)}.

- `limit=1,2,3` for 68%, 95% and 99% Confidence Level results, respectively

Let's try to get the Planck constrtaints on $n_s$

```python
print('Planck constraints at 68 CL:', samples.getInlineLatex('ns',1) )
print('Planck constraints at 95 CL:', samples.getInlineLatex('ns',2) )
print('Planck constraints at 99 CL:', samples.getInlineLatex('ns',3) )
```

### Step 7: Correlation among cosmological parameters (contour plots)

We can figure out the correlation among the different parameters by plotting the marginalized contours that define the 68% and 95% confidence level regions in a 2D plane. This allows us to visualize where it is more probable to find the parameters and understand their correlation:


Let's try for ($n_s$, $H_0$)

```python
g = plots.get_single_plotter(width_inch=3.5,ratio=1)

g.plots_2d(samples, ['ns', 'H0'], filled=True,)
plt.grid(lw=0.4,ls='--')
```

You can use 
- `settings.num_plot_contours = 1` to plot only plot the 68% CL
- `settings.num_plot_contours = 2` to plot only plot the 68 and 95% CL
- `settings.num_plot_contours = 3` to plot only plot the 68, 95% and 99% CL

```python
g = plots.get_single_plotter(width_inch=3.5,ratio=1)

g.settings.num_plot_contours = 3

g.plot_2d(samples, 'ns', 'H0', filled=True);
plt.grid(lw=0.4,ls='--')
```

You can even choose the probability contour yourself when uploading the chains (should be specified in `settings`).

For instance, to plot the contours at 20%, 40%, 60%, and 80% confidence levels, you can do:    

```python
new_samples = getdist.loadMCSamples(path_to_chain, 
                                settings={'ignore_rows':0.3, 
                                          #choose here your favourite contours
                                          'contours':[0.2, 0.4, 0.6, 0.8]} 
                               )

g = plots.get_single_plotter(width_inch=3.5,ratio=1)
g.settings.num_plot_contours = 4
g.plots_2d(new_samples, ['ns', 'H0'], filled=True,)
plt.grid(lw=0.4,ls='--')
```

### Step 10: Reweithing Sample

We can do importance sampling, modifying the samples by re-weighting by a new likelihood

For instance, we can focus on the results obtained within $w$CDM model of cosmology where Dark Energy is assumed to be a fluid with equation of state $w$ free to vary in the MCMC.

We upload the chains:

```python
path_to_wcdm='../data/Planck_chains/wcdm/wcdm'

samples_wcdm = getdist.loadMCSamples(path_to_wcdm, 
                                settings={'ignore_rows':0.3} 
                               )
```

Focusing on CMb only data, in this model $H_0$ is essentially unconstrained due to the degeneracy with $w$

```python
g = plots.get_single_plotter(width_inch=3.5,ratio=1)

g.plots_2d(samples_wcdm, ['w', 'H0'], filled=True,)
plt.grid(lw=0.4,ls='--')
```

```python
print(samples_wcdm.getInlineLatex('H0',2), 'at 95%CL')
print(samples_wcdm.getInlineLatex('w',2), 'at 95%CL')
```

We can assume a prior $H_0=73.04\pm 1.04$ km/s/Mpc from SH0ES and see what happens

```python
#Make a copy of the samples:
rew_samples = samples_wcdm.copy() # make a copy so don't change the original

#Define the reweithed loglike
rew_loglike = (samples_wcdm['H0']-73.04)**2/1.04**2/2

#Add loglike to the Reweithing Sample
rew_samples.loglikes = np.zeros(samples_wcdm.numrows) 
rew_samples.reweightAddingLogLikes(rew_loglike) 
```

Let's check the convergence of the reweighted samples and the new results for $w$ and $H_0$

```python
print('R-1=',rew_samples.getGelmanRubin(),'\n')
print(rew_samples.getInlineLatex('H0',1), 'at 68%CL')
print(rew_samples.getInlineLatex('w',1), 'at 68%CL')
```

Let's compare the best-fit $\chi^2$ in the two cases:

```python
samples_wcdm['chi2'].min()
```

```python
rew_samples['chi2'].min()
```

```python
g = plots.get_single_plotter(width_inch=3.5,ratio=1)
g.plot_2d([samples_wcdm,rew_samples], 'w','H0', filled=True)
plt.grid(lw=0.4,ls='--')
g.add_legend([r'Planck-2018',r'Planck-2018+SH0ES'],legend_loc='lower right',fontsize=10)
```

### 11 Define new parameters

Starting from the parameters sampled in the chains, you can define derived parameters where needed.

Here, as an example, we derive the matter cluster parameter \( S_8 = \sigma_8 \sqrt{\frac{\Omega_m}{0.3}} \).

- $\Omega_m $ (i.e., the matter density at present) is already contained in the chains.
- $ \sigma_8$ (i.e, the root mean square density fluctuation in spheres of 8 Mpc/h radius) is also in the chains.
- $ S_8 $ is not defined yet.

```python
g = plots.get_single_plotter(width_inch=3.5,ratio=1)
g.plot_2d([samples_wcdm,rew_samples], 'omegam','sigma8', filled=True)
plt.grid(lw=0.4,ls='--')
g.add_legend([r'Planck-2018',r'Planck-2018+SH0ES'],legend_loc='upper right',fontsize=10)
```

Let's define $S_8$ for Planck-2018... 

```python
p = samples_wcdm.getParams() 
p.S8 =samples_wcdm['sigma8'] * (samples_wcdm['omegam'] /0.3)
samples_wcdm.addDerived(p.S8, name='S8', label='S_8');
```

... and for Planck-2018+SH0ES

```python
p = rew_samples.getParams() 
p.S8 =rew_samples['sigma8'] * (rew_samples['omegam'] /0.3)
rew_samples.addDerived(p.S8, name='S8', label='S_8');
```

Let's plot them:

```python
g = plots.get_single_plotter(width_inch=3.5,ratio=1)
g.plot_2d([samples_wcdm,rew_samples], 'omegam','S8', filled=True)
plt.grid(lw=0.4,ls='--')
g.add_legend([r'Planck-2018',r'Planck-2018+SH0ES'],legend_loc='upper right',fontsize=10)
```

Let's plot them in 3D, as well

```python
g = plots.get_subplot_plotter(width_inch=4.5)
g.plots_3d(rew_samples,[['omegam', 'S8','sigma8']]);
g.add_2d_contours(rew_samples, 'omegam','S8',ls='--',lw=0.7);
plt.grid(lw=0.4,ls='--')
g.add_legend([r'Planck-2018+SH0ES'],legend_loc='upper left',fontsize=10)
```
